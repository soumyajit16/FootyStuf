```python
import matplotlib.patheffects as path_effects
import matplotlib.pyplot as plt
import pandas as pd
from mplsoccer import Pitch, FontManager, add_image
```

```python
URL = 'https://fbref.com/en/share/3rfMI'
df_2021 = pd.read_html(URL)[0]
df_2021=df_2021.dropna()
df_2021.reset_index(drop=True, inplace = True)


df_2021.set_axis(['Stat','20-21 per 90', '20-21 Percentile'], axis=1, inplace=True)
df_2021=df_2021.set_index('Stat')
df_2021.to_csv('D:\Juego\dejong_2021.csv')

df_2021
```

```python
URL = 'https://fbref.com/en/share/C3F55'

df_1920 = pd.read_html(URL)[0]
df_1920=df_1920.dropna()
df_1920.reset_index(drop=True, inplace = True)
df_1920.set_axis(['Stat','19-20 per 90', '19-20 Percentile'], axis=1, inplace=True)
df_1920=df_1920.set_index('Stat')
df_1920.to_csv('D:\Juego\dejong_1920.csv')
df_1920
```

```python
df= pd.concat([df_2021, df_1920], axis=1)
df['20-21 per 90'] = df['20-21 per 90'].str.rstrip('%').astype('float') / 1.0
df['19-20 per 90'] = df['19-20 per 90'].str.rstrip('%').astype('float') / 1.0
df
```

```python
df=df.dropna()
df=df.drop(['xG',
 'xA','Average Shot Distance',
 'Passes Completed',
 'Passes Attempted','Total Passing Distance',
 'Progressive Passing Distance',
 'Key Passes',
 'Passes into Final Third','Live-ball passes',
 'Through Balls','Ground passes',
 'Shot-Creating Actions',
 'Tackles','Dribbled Past','Successful Pressure %','Progressive Carrying Distance',
 'Carries into Final Third',
 'Dispossessed'])
df
```

# 1. per 90 stats comparison

### Pizza Chart

```python
values = df['19-20 per 90'].tolist() #19-20 szn
values_2 = df['20-21 per 90'].tolist()  #20-21 szn stats
```

```python
import matplotlib.pyplot as plt
from highlight_text import fig_text
from mplsoccer import PyPizza, FontManager
font_normal = FontManager(("https://github.com/google/fonts/blob/main/apache/roboto/static/"
                           "Roboto-Regular.ttf?raw=true"))
font_italic = FontManager(("https://github.com/google/fonts/blob/main/apache/roboto/static/"
                           "Roboto-Italic.ttf?raw=true"))
font_bold = FontManager(("https://github.com/google/fonts/blob/main/apache/roboto/static/"
                         "Roboto-Medium.ttf?raw=true"))
```

```python
# instantiate PyPizza class
baker = PyPizza(
    params=params,# list of parameters
    min_range=min_range,        # min range values
    max_range=max_range,        # max range values
    background_color="#EBEBE9",     # background color
    straight_line_color="#222222",  # color for straight lines
    straight_line_lw=1,             # linewidth for straight lines
    last_circle_lw=1,               # linewidth of last circle
    last_circle_color="#222222",    # color of last circle
    other_circle_ls="-.",           # linestyle for other circles
    other_circle_lw=1               # linewidth for other circles
)

# plot pizza
fig, ax = baker.make_pizza(
    values,                     # list of values
    compare_values=values_2,    # comparison values
    figsize=(15, 15),color_blank_space="same",   # use same color to fill blank space
    blank_alpha=0.3,            # adjust figsize according to your need
    kwargs_slices=dict(
        facecolor="#1A78CF", edgecolor="#222222",
        zorder=2, linewidth=1
    ),                          # values to be used when plotting slices
    kwargs_compare=dict(
        facecolor="#FF9300", edgecolor="#222222",
        zorder=2, linewidth=1,
    ),
    kwargs_params=dict(
        color="#000000", fontsize=12,
        fontproperties=font_normal.prop, va="center"
    ),                          # values to be used when adding parameter
    kwargs_values=dict(
        color="#000000", fontsize=12,
        fontproperties=font_normal.prop, zorder=3,
        bbox=dict(
            edgecolor="#000000", facecolor="cornflowerblue",
            boxstyle="round,pad=0.2", lw=1
        )
    ),                          # values to be used when adding parameter-values labels
    kwargs_compare_values=dict(
        color="#000000", fontsize=12, fontproperties=font_normal.prop, zorder=3,
        bbox=dict(edgecolor="#000000", facecolor="#FF9300", boxstyle="round,pad=0.2", lw=1)
    ),                          # values to be used when adding parameter-values labels
)

# add title
fig_text(
    0.515, 0.99, "<Alexia Putellas> vs <League Average>",
    size=16, fig=fig,
    highlight_textprops=[{"color": '#1A78CF'}, {"color": '#FF9300'}],
    ha="center", fontproperties=font_bold.prop, color="#F2F2F2"
)

# add subtitle
fig.text(
    0.515, 0.955,
    "Frenkie De Jong in numbers (per 90 in La Liga) | 19/20 vs 2021",
    size=15,
    ha="center", fontproperties=font_bold.prop, color="#000000"
)

# add credits
CREDIT_1 = "data: statsbomb viz fbref"
#CREDIT_2 = "inspired by: @Worville, @FootballSlices, @somazerofc & @Soumyaj15209314"

fig.text(
    0.99, 0.005, f"{CREDIT_1}", size=9,
    fontproperties=font_italic.prop, color="#000000",
    ha="right"
)


# add text
fig.text(
    0.43, 0.93, "19-20                                            20-21       ", size=14,
    fontproperties=font_bold.prop, color="#370f59"
)

# add rectangles
fig.patches.extend([
    plt.Rectangle(
        (0.40, 0.9225), 0.025, 0.021, fill=True, color="#1a78cf",
        transform=fig.transFigure, figure=fig
    ),
    plt.Rectangle(
        (0.60, 0.9225), 0.025, 0.021, fill=True, color="#ff9300",
        transform=fig.transFigure, figure=fig
    )])

plt.show()
```

### Radar Chart

```python
val=[values,values_2]
ranges=[]

max_range=[0.15,0.15,0.75,95.0,5.5,14,1.2,1.0,13.5,0.15,70]
min_range= [0.05,0.05,0.05,0.05,0.05,0.0,0.05,0.05,0.05,0.0,0.05]

for x in range(len(params)):
    
    
    ranges.append((min_range[x],max_range[x]))
    
ranges
```

```python
from soccerplots.radar_chart import Radar

title = dict(
    title_name='FdJ in La Liga 19/20',
    title_color = 'red',
    subtitle_name = '(stats per 90)',
    subtitle_color = 'red',
    title_name_2='FdJ in La Liga 20/21',
    title_color_2 = 'blue',
    subtitle_name_2 = '(stats per 90)',
    subtitle_color_2 = 'blue',
    title_fontsize = 18,
    subtitle_fontsize=15
)
endnote = 'via FBREF / Statsbomb'

radar = Radar()

fig,ax = radar.plot_radar(ranges=ranges,params=params,values=val,
                         radar_color=['red','blue'],
                         alphas=[.75,.6],title=title,endnote=endnote,
                         compare=True)
fig.savefig('D:\Juego\ComparingDeJong.jpg')
```
